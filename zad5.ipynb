```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Funkcja wielomianowa do dopasowania
def polynomial(x, a, b, c, d, e):
    return a*x**4 + b*x**3 + c*x**2 + d*x + e

# Zakres kątów od 0 do 2*pi
theta = np.linspace(0, 2*np.pi, 1000)

# Równanie krzywej w współrzędnych biegunowych
r = np.sin(2 * theta)

# Przekształcenie współrzędnych biegunowych na kartezjańskie
x = r * np.cos(theta)
y = r * np.sin(theta)

# Dopasowanie wielomianu do danych krzywej
popt, _ = curve_fit(polynomial, x, y)

# Narysowanie wykresu krzywej i dopasowanego wielomianu
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(x, y, label='Krzywa czterolistna: r(θ) = sin(2θ)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Wykres krzywej czterolistnej')
plt.grid(True)
plt.axis('equal')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x, y, label='Krzywa czterolistna: r(θ) = sin(2θ)')
plt.plot(x, polynomial(x, *popt), label='Dopasowany wielomian')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Dopasowany wielomian')
plt.grid(True)
plt.axis('equal')
plt.legend()

plt.tight_layout()
plt.show()

# Wyświetlenie współczynników wielomianu
print("Współczynniki wielomianu:", popt)
```
