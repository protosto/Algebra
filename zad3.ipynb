```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Tworzenie siatki punktw w przestrzeni trjwymiarowej
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
x, y = np.meshgrid(x, y)

# Rozmaito 1: V (z  x^2  y^2)
z1 = x**2 + y**2

# Rozmaito 2: V (z^2  x^2  y^2)
z2 = np.sqrt(x**2 + y**2)

# Rozmaito 3: V (z  x^2 + y^2)
z3 = x**2 - y**2

# Rozmaito 4: V (xz, yz)
z4 = x * y

# Wykresy
fig = plt.figure(figsize=(15, 10))

# Rozmaito 1
ax1 = fig.add_subplot(221, projection='3d')
ax1.plot_surface(x, y, z1, cmap='viridis')
ax1.set_title('Rozmaito 1: V (z  x^2  y^2)')

# Rozmaito 2
ax2 = fig.add_subplot(222, projection='3d')
ax2.plot_surface(x, y, z2, cmap='plasma')
ax2.set_title('Rozmaito 2: V (z^2  x^2  y^2)')

# Rozmaito 3
ax3 = fig.add_subplot(223, projection='3d')
ax3.plot_surface(x, y, z3, cmap='inferno')
ax3.set_title('Rozmaito 3: V (z  x^2 + y^2)')

# Rozmaito 4
ax4 = fig.add_subplot(224, projection='3d')
ax4.plot_surface(x, y, z4, cmap='magma')
ax4.set_title('Rozmaito 4: V (xz, yz)')

plt.show()
```
